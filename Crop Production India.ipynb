```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from category_encoders import TargetEncoder
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LabelEncoder()
```

```python
ce = TargetEncoder()
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer()
```

```python
num_pipeline1 = Pipeline(
    steps = [('replace_missing_values',KNNImputer()),
            ('standscalar',StandardScaler())]
)
```

```python
num_pipeline1
```

```python
cat_pipe1 = Pipeline(
    steps = [('Target encoding',ce),
            ('standscalar',StandardScaler())]
)
```

```python
cat_pipe1
```

```python
process_1 = ColumnTransformer(
    [('numerical_variable',num_pipeline1,x.select_dtypes(exclude='object').columns),
    ('categorical variable',cat_pipe1,x.select_dtypes(include='object').columns)]
)
```

```python
process_1
```

```python
from sklearn.pipeline import make_pipeline
pipe_1 = make_pipeline(process_1,RandomForestRegressor())

pipe_2 = make_pipeline(process_1,DecisionTreeRegressor())

pipe_3 = make_pipeline(process_1, XGBRegressor() )

pipe_4 = make_pipeline(process_1, CatBoostRegressor() )

pipe_5 = make_pipeline(process_1, GradientBoostingRegressor() )
```

```python
pipe_1
```

```python
pipes = [pipe_1,pipe_2,pipe_3,pipe_4,pipe_5]
```

```python
model = ['Random_forest','Decision_tree','XGB','Catboost','Gradient']
```

```python
dict1 = dict(zip(pipes,model))
```

```python
for pipe in pipes:
    pipe.fit(x_train,y_train)
```

```python
for pipe in pipes:
    print(f'{dict1[pipe]} score is {pipe.score(x_train,y_train)}')
```

```python
for pipe in pipes:
    print(f'{dict1[pipe]} score is {pipe.score(x_test,y_test)}')
```
